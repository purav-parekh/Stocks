```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2

figure()
plot(x,y,'r')
xlabel('x')
ylabel('y')
title('title')
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('title');
```

```python
img = cv2.imread('rajinikanth-759.jpg', 0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)
```

```python
img = cv2.imread('rajinikanth-759.jpg',0)
rows,cols = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)
cv2.waitKey(0)
```
